### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
import pandas as pd
import numpy as np

def perform_statistical_analysis(url: str) -> pd.DataFrame:
    """
    Carga un conjunto de datos y calcula estadísticas descriptivas clave 
    para las columnas numéricas.
    """
    # 1. Carga del dataset
    df = pd.read_csv(url)
    
    # 2. Selección de columnas numéricas
    # Excluimos columnas que actúan como ID (como 'Unnamed: 0' o 'Customer')
    numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if 'Unnamed: 0' in numerical_cols:
        numerical_cols.remove('Unnamed: 0')

    # 3. Cálculo de estadísticas
    # .describe() nos da media, desviación estándar, cuartiles, min y max
    stats = df[numerical_cols].describe().T
    
    # 4. Añadimos la Mediana y la Moda (que no vienen por defecto en .describe)
    stats['median'] = df[numerical_cols].median()
    
    # La moda puede devolver múltiples valores, tomamos el primero
    stats['mode'] = df[numerical_cols].mode().iloc[0]
    
    # Reordenamos las columnas para mayor claridad
    columns_order = ['mean', 'median', 'mode', 'std', 'min', '25%', '50%', '75%', 'max']
    return stats[columns_order]
stats_results = perform_statistical_analysis("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
print(stats_results)
```

Statistical Distribution & Business Insights
1. Income: The Zero-Value Segment

Standard Deviation Volatility: The Std Dev ($1,601) is much higher than the Mean ($193). In a normal distribution, this would be impossible; here, it confirms that the extreme outliers are heavily distorting the "average" cost of a policy.

Standard Market: 75% of your customers pay $109.50 or less. This shows the core business is volume-based, focusing on affordable, standard insurance.

3. Total Claim Amount: Risk Concentration
Predictable Claims: Unlike premiums, claims are more "stable." The Median ($354) and Mean ($404) are relatively close, meaning the distribution is less distorted by extreme values.

The Profitability Buffer: The 75th percentile for claims is $532, while 75% of premiums are only $109. This highlights the insurance fundamental: it takes roughly 5 safe customers paying their monthly premium to cover the cost of 1 average claim.

Maximum Loss: Even the worst-case claim in this set ($2,893) is manageable, as it doesn't reach the extreme levels seen in the premium outliers.

The marketing team wants to know the top 5 less common customer locations. Create a pandas Series object that contains the customer locations and their frequencies, and then retrieve the top 5 less common locations in ascending order.

```python
import pandas as pd

# 1. Create the Series with frequencies
# We use value_counts() which returns a Series indexed by location
location_counts = df['ST'].value_counts()

# 2. Retrieve the top 5 less common locations
# We sort in ascending order (default) to get the smallest counts first
bottom_5_locations = location_counts.sort_values().head(5)

# Display the result
print("Top 5 less common customer locations:")
print(bottom_5_locations)
```

The sales team wants to know the total number of policies sold for each type of policy. Create a pandas Series object that contains the policy types and their total number of policies sold, and then retrieve the policy type with the highest number of policies sold.

*Hint:*
- *Using value_counts() method simplifies this analysis.*
- *Futhermore, there is a method that returns the index of the maximum value in a column or row.*


```python
import pandas as pd

# 1. Create a pandas Series with the counts of each policy type
policy_counts = df['Policy Type'].value_counts()

# 2. Retrieve the policy type with the highest number of policies sold
# Since value_counts() sorts by descending order, the first index [0] is the highest
highest_policy_type = policy_counts.idxmax()
highest_value = policy_counts.max()

# Display the results
print("Total policies sold by type:")
print(policy_counts)
print(f"\nThe policy type with the highest sales is: {highest_policy_type} ({highest_value} policies)")
```

### Exercise 3

The sales team wants to know if customers with Personal Auto have a lower income than those with Corporate Auto. How does the average income compare between the two policy types?

- Use *loc* to create two dataframes: one containing only Personal Auto policies and one containing only Corporate Auto policies.
- Calculate the average income for each policy.
- Print the results.

```python
# Your code hereimport pandas as pd

# 1. Filtramos los datos para obtener solo las filas de 'Personal Auto'
# Luego seleccionamos la columna 'Income'
personal_auto_income = df.loc[df['Policy Type'] == 'Personal Auto', 'Income']

# 2. Filtramos los datos para obtener solo las filas de 'Corporate Auto'
# Luego seleccionamos la columna 'Income'
corporate_auto_income = df.loc[df['Policy Type'] == 'Corporate Auto', 'Income']

# 3. Calculamos el promedio (mean) de cada grupo por separado
mean_personal = personal_auto_income.mean()
mean_corporate = corporate_auto_income.mean()

# 4. Mostramos los resultados
print(f"Average Income (Personal Auto):  ${mean_personal:,.2f}")
print(f"Average Income (Corporate Auto): ${mean_corporate:,.2f}")
```
